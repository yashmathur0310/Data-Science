```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
tips.isnull().sum()
```

```python
tips.corr()
```

```python
sns.heatmap(tips.corr(),annot=True)
```

```python
tips.describe()
```

### Distribution of total bill and tip

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.distplot(tips['total_bill'],bins=10)
plt.subplot(1,2,2)
sns.distplot(tips['tip'],bins=10)
```

### Relation between total bill and tip

```python
sns.jointplot(x='total_bill',y='tip',kind='hex',data=tips)
```

### Analysis of total bill with respect to day 

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.boxplot(x='day',y='total_bill',data=tips)
plt.subplot(1,2,2)
sns.swarmplot(x='day',y='total_bill',data=tips)
```

### Analysis of total bill with respect to day(hue=sex,smoker,time)


```python
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
sns.boxplot(x='day',y='total_bill',hue='sex',data=tips)
plt.subplot(1,3,2)
sns.boxplot(x='day',y='total_bill',hue='smoker',data=tips)
plt.subplot(1,3,3)
sns.boxplot(x='day',y='total_bill',hue='time',data=tips)

```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
sns.swarmplot(x='day',y='total_bill',hue='sex',data=tips)
plt.subplot(1,3,2)
sns.swarmplot(x='day',y='total_bill',hue='smoker',data=tips)
plt.subplot(1,3,3)
sns.swarmplot(x='day',y='total_bill',hue='time',data=tips)
```

### Analysis of tip with respect to day

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.boxplot(x='day',y='tip',data=tips)
plt.subplot(1,2,2)
sns.stripplot(x='day',y='tip',data=tips)
```

### Analysis of tip with respect to day (hue =sex,smoker,time)

```python
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
sns.boxplot(x='day',y='tip',hue='sex',data=tips)
plt.subplot(1,3,2)
sns.boxplot(x='day',y='tip',hue='smoker',data=tips)
plt.subplot(1,3,3)
sns.boxplot(x='day',y='tip',hue='time',data=tips)

```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
sns.stripplot(x='day',y='tip',hue='sex',data=tips)
plt.subplot(1,3,2)
sns.stripplot(x='day',y='tip',hue='smoker',data=tips)
plt.subplot(1,3,3)
sns.stripplot(x='day',y='tip',hue='time',data=tips)

```

1.Tip offered by males is more than females. 

2.Sum of total bill is more of males than females.

3.Lunch is preferred more on thursday , friday and there are no customers in lunch on saturday and sunday.

4.Total bill and tips of non smokers is more than smokers. 

5.Majority of the total bills are in the range of 10-30 and tip offered is in the range of 2-6.

6.Tip offered at Saturday and Sunday is more than friday and thursday.

7.Male guests are more than as compared to females.

8.Sum of total bill as well as tip is the highest on saturday.
