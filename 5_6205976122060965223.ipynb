```python
import pandas as pd 
import numpy as np 
```

```python
df=pd.read_excel('Continent List.xlsx')
df.head()
```

```python
df['Continent'].unique()
```

```python
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
def get_first_alphabet(text):
    text=text.split()[0]
    for alphabet in text:
        return(alphabet[0])
```

```python
df['First_alphabet']=df['Country'].apply(lambda x:get_first_alphabet(x))
```

```python
df_first=pd.DataFrame(df['First_alphabet'].value_counts()).reset_index()
```

```python
df_first.columns=['Alphabet','first_Count']
```

```python
df_first=df_first.sort_values(['Alphabet'],ascending=True).reset_index().drop('index',axis=1)
```

```python
df_first
```

```python
def get_last_alphabet(text):
    text=text.split()
    for alphabet in text:
        return(alphabet[-1])
```

```python
get_last_alphabet("turkey")
```

```python
df['Last_alphabet']=df['Country'].apply(lambda x:get_last_alphabet(x))
```

```python
df_last=pd.DataFrame(df['Last_alphabet'].value_counts()).sort_index(ascending=True).reset_index()
```

```python
df_last.columns=["alphabet",'last_count']
```

```python
df_last
```

```python
final=pd.concat([df_first,df_last],axis=1)
```

```python
final=final.drop("alphabet",axis=1)
```

```python
final=final.fillna(0)
```

```python
final['Sum']=final['first_Count']+final['last_count']
```

```python
final
```

```python
def get_continent(text):
    if text=='Asia':
        return("Asian")
    else:
        return("Non-Asian")
```

```python
df['Countries_list']=df['Continent'].apply(lambda x:get_continent(x))
```
